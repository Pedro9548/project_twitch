```python
# n! = n*(n-1)!

def factorial(n):
    space = ' ' * (4 * n)
    print(space, 'factorial', n)
    if n == 0:
        print(space, 'returning 1')
        return 1
    else:
        recurse = factorial(n-1)
        result = n * recurse
        print(space, 'returning', result)
        return result

factorial(3)
```

```python
def is_reverse(word1, word2):
    if len(word1) != len(word2):
        return False
    i = 0
    j = len(word2) - 1
    while j > 0:
        if word1[i] != word2[j]:
            return False
        i = i+1
        j = j-1
    
    return True


is_reverse("pneu", "pneu1")
        

```

```python
def rotate_word(word, num):

    string_list=[]

    for letters in word:
        converted = ord(letters)+num
        string_list.append(chr(converted))

        teste = ''.join(string_list)

    return teste

rotate_word("HAL", 1)
```
